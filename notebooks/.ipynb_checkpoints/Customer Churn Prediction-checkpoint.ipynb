```python
df.drop(columns='customerID', inplace=True)
```

```python
binary_cols = [
    'Partner','Dependents','PhoneService','PaperlessBilling',
    'TechSupport','StreamingTV','StreamingMovies',
    'OnlineSecurity','OnlineBackup','DeviceProtection'
]

for col in binary_cols: 
    df[col] = df[col].map({'No': 0, 'Yes': 1})
```

```python
df = pd.get_dummies(
    df,
    columns=['Contract','PaymentMethod','InternetService','MultipleLines','gender'],
    drop_first=True
)
```

```python
internet_dependent_cols = [
    'OnlineSecurity','OnlineBackup','DeviceProtection',
    'TechSupport','StreamingTV','StreamingMovies'
]

df[internet_dependent_cols] = df[internet_dependent_cols].fillna(0)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=42, 
    stratify=y,
    test_size=0.2
)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

# Feature Scaling (Standard Scale)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
X_train[num_cols].describe()
```

# Logistic Regression Model (Base Model)

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_LR = LogisticRegression(
    max_iter=1000,
    class_weight='balanced',
    random_state=42
)
```

```python
model_LR.fit(X_train, y_train)
```

```python
y_pred = model_LR.predict(X_test)
```

```python
y_prob = model_LR.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score
)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]
```

```python
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_rf))
```

# Extracting Business Insights from Both Models (LR & RF)

## From Logistic Regression

```python
import pandas as pd

lr_coeff = pd.Series(
    model_LR.coef_[0],
    index=X_train.columns
).sort_values()
```

```python
lr_coeff
```

## From Random Forest

```python
rf_importance = pd.Series(
    rf_model.feature_importances_,
    index=X_train.columns
).sort_values(ascending=False)
```

```python
rf_importance
```

# XG Boost

```python
from xgboost import XGBClassifier
```

```python
xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='binary:logistic',
    eval_metric='auc',
    scale_pos_weight=(y_train.value_counts()[0] / y_train.value_counts()[1]),
    random_state=42
)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test)
y_prob_xgb = xgb_model.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

print(classification_report(y_test, y_pred_xgb))
print(confusion_matrix(y_test, y_pred_xgb))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_xgb))
```
